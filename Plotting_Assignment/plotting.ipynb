```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
sales = pd.DataFrame({
    "Month": months,
    "Revenue": np.random.randint(80,150,12),
    "Expenses": np.random.randint(50,100,12),
    "Region": np.random.choice(["East","West","North","South"],12)
})
sales["Profit"] = sales["Revenue"] - sales["Expenses"]

names = ["Alice","Bob","Charlie","Diana","Evan","Fiona","George","Hannah"]
depts = ["HR","Sales","IT","Marketing"]
perf = pd.DataFrame({
    "Employee": names,
    "Department": np.random.choice(depts,len(names)),
    "Projects": np.random.randint(3,15,len(names)),
    "Hours": np.random.randint(120,200,len(names)),
    "Score": np.random.randint(1,11,len(names))
})

sales, perf

# 1. Line Chart
plt.plot(sales["Month"], sales["Revenue"], marker="o")
plt.title("Revenue Over Time")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()

# 2. Bar Chart
x = np.arange(len(sales["Month"]))
w = 0.4
plt.bar(x - w/2, sales["Revenue"], w, label="Revenue")
plt.bar(x + w/2, sales["Expenses"], w, label="Expenses")
plt.title("Revenue vs Expenses")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.legend()
plt.xticks(x, sales["Month"], rotation=30)
plt.show()

# 3. Area Chart
xp = np.arange(len(sales["Month"])).astype(float)
y = sales["Profit"].astype(float)
plt.plot(xp, y, marker="o")
plt.fill_between(xp, y, 0, alpha=0.3)
plt.title("Profit")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.xticks(xp, sales["Month"], rotation=60)
plt.show()

# 4. Scatter chart
plt.scatter(perf["Hours"], perf["Projects"])
plt.title("Hours vs Projects done")
plt.xlabel("Hours worked")
plt.ylabel("Projects completed")
plt.show()

# 5. Histogram
plt.hist(perf["Score"], bins=5)
plt.title("Satisfaction Scores")
plt.xlabel("Score")
plt.ylabel("Number of employees")
plt.show()

# 6. Pie Chart
c = perf["Department"].value_counts()
plt.pie(c, labels=c.index, autopct="%1.0f%%")
plt.title("Employees by Department")
plt.show()
```
